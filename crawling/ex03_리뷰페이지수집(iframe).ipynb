### iframe
- 화면에는 존재하지만 실제데이터는 다른공간에 존재하는 경우
- 데이터를 수집하기 위해서는 데이터가 존재하는 공간으로 접근
- iframe은 흔치않은 웹페이지 구성(요즘 트렌드에 선호하지 않음)

```python
# 리뷰데이터가 들어있는 페이지 주소로가서 리뷰데이터를 수집
url2 = 'https://review4.cre.ma/bodyluv.kr/mobile/products/reviews?product_code=437&iframe_id=crema-product-reviews-2&widget_id=7&iframe=1&widget_style=&install_method=hardcoded&app=0&device=mobile&parent_url=https%3A%2F%2Fm.bodyluv.kr%2Fproduct%2F%25EB%25B0%2594%25EB%2594%2594%25EB%259F%25BD-%25EB%2594%25A5%25EC%258A%25AC%25EB%25A6%25BD-%25EC%25BF%25A8-%25EC%259D%25B4%25EB%25B6%2588-v2-%25EC%2595%2588%25ED%258B%25B0%25EB%25B2%2584%25EA%25B7%25B8%2F437%2Fcategory%2F1%2Fdisplay%2F2%2F%23prdReview&nonmember_token=&secure_device_token=V221cb166d8de710c3bfcaab5a3abc2954602f0ec97c71a71b7c78e582f801e435524cf89909f124ba2a43d06c06a85ba0'
res2 = req.get(url2)
soup2 = bs(res2.text,'lxml')
review = soup2.select('div.review_list_v2__message.js-collapsed-review-content.js-translate-text')
```

```python
# strip() : 문자열 이외에 불필요한 특수기호(공백, 개행기호 등) 제거하고 출력
review[0].text.strip()
```

```python
for i in range(len(review)):
    print(f'{i+1}번째 리뷰 : ', end="")
    print(review[i].text.strip())
```
