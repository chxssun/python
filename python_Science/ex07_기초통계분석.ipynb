```python
import pandas as pd
```

```python
# hr데이터를 활용하여 데이터 탐색
# hr데이터 불러오기
hr_info = pd.read_csv('data/hr-info.csv', encoding='euc-kr')
hr_info.head()
```

```python
# 기술통계량 확인 -> 수치데이터의 기본적인 통계수치를 보여준다
hr_info.describe()
# count : 데이터의 개수
# mean : 평균
# std : 표준편차(평균에서 값들이 떨어져있는 정도) -> 분산에 루트를 씌운
# min : 최솟값, max L: 촤댓값
# 25%, 50%, 75% : 사분위수 
    # -> 데이터를 오름차순 정렬 후 순서대로 25%,50%,75% 위치해 있는값
```

```python
hr_info.info()
```

```python
# object(문자열) 타입의 기술통계량 확인
hr_info.describe(include=['object'])
# count : 데이터의 개수
# unique : 중복 없이 유일한 값의 개수
# top : 유일한 값들 -> 카테고리들 중 가장 많이 등장한 값
# freq : frequency빈도수 -> top 값이 나온 빈도수
```

### 범용함수
- 데이터.sum() : 합계를 구할 때 사용
- 데이터.mean() : 평균을 구할 때 사용
- 데이터.count() : 데이터의 개수를 셀때 사용 (값이 있으면 카운팅)
- 데이터.value_counts() : 각 값의 개수를 셀때 사용 
- 데이터.unique() : 중복없이 유일한 값들을 출력할 때 사용
- 데이터.min() : 최솟값
- 데이터.max() : 최댓값

```python
# 성별데이터의 개수 확인 -> count()
hr_info['GENDER']
hr_info['GENDER'].count()
```

```python
# 성별데이터 값의 개수 확인 -> value_counts()
hr_info['GENDER'].value_counts()
```

```python
# 직원들의 국적 종류를 알아보자
hr_info['NATION'].unique()
```

```python
# 최대근속연수, 최소 근속년수 확인
print(f'최대 근속연수 : {hr_info["SERVICE_YEAR"].max()}')
print(f'최소 근속연수 : {hr_info["SERVICE_YEAR"].min()}')
```

```python
# 평균근속년수, 중앙값을 활용한 근속연수
print(f'평균 근속년수 : {hr_info['SERVICE_YEAR'].mean()}')
print(f'중앙값 근속년수 : {hr_info['SERVICE_YEAR'].median()}')
```

- 교육수준에 따른 평균 연봉 확인

```python
# 교육수준의 종류 확인
hr_info['EDU_LEVEL'].unique()
```

```python
# 교육수준이 '박사학위' 인 직원의 평균 평균연봉 확안
# 연봉 : salaey컬럼
edu = hr_info[hr_info['EDU_LEVEL'] == '박사 학위']
edu['SALARY'].mean()
```

```python
hr_info[hr_info['EDU_LEVEL'] == '박사 학위']['SALARY'].mean()
```

```python
# 반복문을 활용해서 전체 교육수준에 따른 연봉 출력
for edu_level in hr_info['EDU_LEVEL'].unique() :
    print(edu_level,hr_info[hr_info['EDU_LEVEL'] == edu_level]['SALARY'].mean())
```

### 데이터 그룹화를 통한 통계분석

```python
# 연봉, 교육수준에 따른 평균값
hr_info[['SALARY',"EDU_LEVEL"]].groupby(by='EDU_LEVEL').mean()
```

```python
# 두개 이상의 기준을 가지고 그룹화
# 교육수준의 및 성별에 따른 평균연봉 확인
# 3개 컬럼 추출
hr_info[['SALARY','EDU_LEVEL','GENDER']]
# 그룹화
hr_info[['SALARY','EDU_LEVEL','GENDER']].groupby(by = ["EDU_LEVEL", "GENDER"]).mean()
```

```python
# 교육수준별 평균 나이 구하기
hr_info['AGE'].groupby(by = hr_info['EDU_LEVEL']).mean()
hr_info[['AGE', 'EDU_LEVEL']].groupby(by = 'EDU_LEVEL').mean()
```

```python
hr_info[['AGE', 'EDU_LEVEL']].groupby(by = 'EDU_LEVEL')
# 두 개의 결과값을 출력
# 그룹화한 카테고리 이름, 그룹화한 결과(DataFrame)
for idx, df in hr_info[['AGE', 'EDU_LEVEL']].groupby(by = 'EDU_LEVEL') :
    display(idx)
    display(df)
```

### 그룹화 후 여러가지의 통계값 확인
- agg()

```python
hr_info[['SALARY', 'EDU_LEVEL']].groupby(by = 'EDU_LEVEL').agg(['mean', 'max','min','median'])
```

### 그룹화 후 컬럼별로 다른 연산결과 확인

```python
# 국적에 따른 평균 성과평과점수와 최대연봉 확인하기
# PERFORMANCE_SCORE -> 평균 mean
# SALARY -> 최댓값 max
hr_info[['NATION', 'PERFORMANCE_SCORE', 'SALARY']].groupby(by = 'NATION').agg(['mean','max'])
hr_info.groupby(by='NATION').agg({'PERFORMANCE_SCORE' : 'mean',
                                    'SALARY' :'max'})
```

### 상관계수 분석
- 두 변수 사이의 관계를 수치화하여 확인하는 방법
- -1 ~ 1 사이의 값을 가진다
- -1 에 가까울수록 음의 상관관계
    - 예시 : 다이어트 의지(x)가 커질수록 체중(y) 감소
- 1 에 가까울수록 양의 상관관계
    - 예시 : 식욕(x)이 커질수록 체중(y) 증가
- 주의! 음수값이라고 해서 상관관계가 없다는 것이 아니다
- 0 에 가까울수록 관련이 없는 변수들이다

```python
# 직원의 성과평가점수와 연봉의 상관계수 확인
hr_info[['PERFORMANCE_SCORE','SALARY']].corr()
# 성과평가점수가 높아질수록 연봉도 높아지는구나
```

```python
# 전체데이터의 상관계수 확인
hr_info.corr(numeric_only=True)
# 절대값으로 1에 가까울수록 상관관계가 높은 변수들이다
# 정답에 영향을 많이 미치는 특성만을 학습시킬때 사용하면 유용
```
