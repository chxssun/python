### 문자열

```python
# 파이썬에서는 문자열에 사용하는 기호로 작은따옴표, 큰따옴표 모두 사용가능
s = 'My name is CS'
print(s)
s1 = "My name is CS"
print(s1)
s2 = "My \nname \nis \nCS"
print(s2)
```

### 문자열의 인덱싱과 슬라이싱 ⭐
- 인덱스 : 색인, 목차 -> 순서가 있다는 뜻
- 인덱싱 : 무엇인가를 가리킨다! 
- 슬라이싱 : 범위를 지정해서 어디서부터 어디까지 잘라낸다

```python
# 인덱싱 -> 기호 : 대괄호
s = 'my name is sun'
print(s[0])
```

```python
# 슬라이싱 -> 기호 : 대활호, 콜론
# 'my' , 'name'문자열 가저오기
s  = 'my name is sun'
print(s[0:2])
print(s[3:7])
```

```python
# 'my name'가져오기
print(s[0:7])
print(s[:7])
print(s[8:])
```

### 문자열 특정 부분 초기화(formatting)

```python
# f문자열 포매팅
month = 8
day = 5
s = f'오늘은 {month}월 {day}일 입니다'
print(s)
```
