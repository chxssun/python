```python
def divisor(num1) :
    for i in range(1, num1+1) :
        if num1 % i == 0 :
            print(i, end=' ')
```

```python
divisor(10)
```

### 딕셔너리 형태의 가변 매개변수(**)
- 딕셔너리 형태로 함수 내부를 처리하고 싶을때

```python
def print_map(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(f'{k}: {v}')
```

```python
maap = print_map(하나 = 1, 둘 = 2)
```

#### lambda함수
- lambda는 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 함
- 람다라고 읽고 익명의 함수하고도 함
- 간단하게 정의할 필요가 있는 함수를 쓸때 사용
- return 명령어가 필요 없다 -> 표현식이 마무리되면 결과값을 돌려주는 함수

```python
g = lambda x: x*2
g(3)
```

```python
# 매개변수가 여러개 일때
h = lambda a,b,c : (a+b)*c
h(1,2,3)
```

## 로또게임 만들기

```python
import random as rd
lotto = []
while True:
    num = rd.randint(1, 45)
    if num not in lotto:
        lotto.append(num)
        if len(lotto) == 6:
            break
lotto
```

```python
inputStr = input("내 로또번호를 6개 입력 : ").split(" ")
mine = []
for n in inputStr:
    mine.append(int(n))
```

```python
same = 0
for myN in mine:
    if myN in lotto:
        same += 1
```

```python
score = [0,6,5,4,3,2,1]
if same == 0:
    print('탈락')
else :
    print(f'{same}개 맞음, {score[same]}등 당첨')
```

```python
import numpy as np
def make_lotto(count):
    for i in range(count):
        lotto_num = []
        
        for j in range(6): 
            lotto_num = np.random.choice(range(1,46), 6, replace=False)
        lotto_num.sort()
        print(f"{i+1}. 로또번호 : {lotto_num}")
```

```python
count = int(input("로또 번호를 몇개 생성할까요? >> "))
make_lotto(count)
```
